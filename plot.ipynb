```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Data
normalized_accuracy = [0.88, 0.4712, 0.2665, 0.264]
size_mb = [117760, 87241.66, 56233.05, 51336.95]
execution_time_sec = [5000, 3698.3, 2358, 2029]

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the data
ax.scatter(normalized_accuracy, size_mb, execution_time_sec, c='r', marker='o')

# Highlight the points by adding lines to the axes
for x, y, z in zip(normalized_accuracy, size_mb, execution_time_sec):
    ax.plot([x, x], [y, y], [0, z], color='gray', linestyle='dashed')

# Connect the points with lines
ax.plot(normalized_accuracy, size_mb, execution_time_sec, color='blue', linestyle='solid')

# Set labels
ax.set_xlabel('Normalized Accuracy')
ax.set_ylabel('Size (MB)')
ax.set_zlabel('Execution Time (sec)')

# Set title
ax.set_title('3D Plot of Normalized Accuracy, Size, and Execution Time')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the exponential growth curve
x = np.linspace(0, 10, 500)
y = np.exp(x / 2) / 100

# Plot the exponential growth curve
plt.plot(x, y, label='Exponential Growth Curve', color='black')

# Add dashed lines for different levels
plt.axhline(0.2, color='gray', linestyle='--', label='Artificial Narrow Intelligence (ANI)')
plt.axhline(1, color='gray', linestyle='--', label='Artificial General Intelligence (AGI)')
plt.axhline(5, color='gray', linestyle='--', label='Artificial Superintelligence (ASI)')
plt.axhline(20, color='gray', linestyle='--', label='Artificial Universal Intelligence (AUI)')

# Annotate each level
plt.text(0.5, 0.22, 'Artificial Narrow Intelligence (ANI)', fontsize=10, verticalalignment='bottom')
plt.text(0.5, 1.02, 'Artificial General Intelligence (AGI)', fontsize=10, verticalalignment='bottom')
plt.text(0.5, 5.02, 'Artificial Superintelligence (ASI)', fontsize=10, verticalalignment='bottom')
plt.text(0.5, 20.02, 'Artificial Universal Intelligence (AUI)', fontsize=10, verticalalignment='bottom')

# Add labels and title
plt.title('The Growth of Artificial Intelligence', fontsize=14)
plt.xlabel('Time', fontsize=12)
plt.ylabel('Capabilities', fontsize=12)

# Remove top and right spines for a clean look
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Adjust the y-axis to log scale for an exponential look
plt.yscale('log')
plt.ylim(0.1, 30)

# Show grid and legend
plt.grid(alpha=0.3)
plt.legend(loc='upper left')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import ast  # To safely evaluate string lists

# List of file paths to process
file_paths = [
    "./final_model/40layers70B/logs/extract_PSO.csv",
    './final_model/30layers70B/logs/extract_PSO.csv',
    './final_model/20layers70B/logs/extract_PSO.csv',
    './final_model/10layers70B/logs/extract_PSO.csv',
]

# Initialize an empty list to collect all x values
all_x_values = []

# Process each file
for file_path in file_paths:
    # Read the CSV file
    data = pd.read_csv(file_path)
    
    # Parse the 'x' column (assuming it's stored as strings of lists)
    for x_list in data['x']:
        x_values = ast.literal_eval(x_list)  # Safely evaluate the string to a list
        all_x_values.extend(x_values)  # Add to the combined list

# Plot the histogram
plt.hist(all_x_values, bins=79, color='C0', zorder = 3)
# plt.title("Histogram of x Values from All Files")
plt.xlabel("The index of pruned layer")
plt.ylabel("Frequency")
plt.grid(zorder = 0)
# plt.show()
plt.savefig('./pso_histogram.pdf',bbox_inches='tight')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import ast  # To safely evaluate string lists

# File paths
file_path1 = "./final_model/10layers70B/logs/extract_PSO.csv"
file_path2 = "./final_model/20layers70B/logs/extract_PSO.csv"
file_path3 = './final_model/30layers70B/logs/extract_PSO.csv'
file_path4 = './final_model/40layers70B/logs/extract_PSO.csv' 

# Load the data
data1 = pd.read_csv(file_path1)
data2 = pd.read_csv(file_path2)
data3 = pd.read_csv(file_path3)
data4 = pd.read_csv(file_path4)

# Extract accuracy values
accuracy1 = np.array(data1['Normalized Accuracy'])*100
accuracy2 = np.array(data2['Normalized Accuracy'])*100
accuracy3 = np.array(data3['Normalized Accuracy'])*100
accuracy4 = np.array(data4['Normalized Accuracy'])*100

# Plot the accuracy line comparison
# plt.figure(figsize=(12, 6))
plt.plot(accuracy1, label='10', marker='o', linestyle='solid', alpha=0.8)
plt.plot(accuracy2, label='20', marker='s', linestyle='dashed', alpha=0.8)
plt.plot(accuracy3, label='30', marker='x', linestyle='dotted', alpha=0.8)
plt.plot(accuracy4, label='40', marker='d', linestyle='dashdot', alpha=0.8)
# plt.title("Comparison of Normalized Accuracy")
plt.xlabel("Steps")
plt.ylabel("Normalized Accuracy [%]")
plt.legend(title='The number of pruned layers',bbox_to_anchor=(0,1.02,1,0.2),loc="lower left",mode='expand',ncol=4)
# plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.grid()
# plt.show()
plt.savefig('./pso_accuracy.pdf',bbox_inches='tight')
```
